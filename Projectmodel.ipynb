```python
def capacity():
  data = pd.read_csv('ticketcount.csv')
  buscount = pd.read_csv('/content/buscount.csv')

  #Count of ticket and route to a dictionary
  fullcount = {}
  for i in range(len(data)):
    f = data['From_loc'][i]+" "+data['To_loc'][i]
    if f not in fullcount:
      fullcount[f] = data['number_of_tickets'][i]
    else:
      fullcount[f]+=data['number_of_tickets'][i]
  #print(fullcount)

  #Dictionary with route name and capacity of passengers
  buscap = {}
  for i in fullcount:
    k = 0
    for j in buscount['Route']:
      k+=1
      if i == j:
        buscap[i] = buscount['Count'][k]*40
  #print(buscap)

  column_names = ["Route", "Exceed_Diminish", "Count_exceeded_diminished"]
  cap = pd.DataFrame(columns = column_names)
  #print(cap)

  #Capacity exceeded or diminished
  for i in buscap:
    if buscap[i] < fullcount[i]:
      print("Exceeded passenger Count by ", fullcount[i]-buscap[i]," in ",i," route")
    else:
      if fullcount[i] < buscap[i]*0.2 :
        print("Passenger Count below by ", buscap[i]-fullcount[i] ," in ",i," route")
  #File containing passenger fall or rise rate
  k=0
  for i in buscap:
    if buscap[i] < fullcount[i]:
      cap.at[k,'Route'] = i
      cap.at[k,"Exceed_Diminish"] = 1
      cap.at[k,"Count_exceeded_diminished"] = fullcount[i]-buscap[i]
      k+=1
    else:
      if fullcount[i] < buscap[i]*0.2 :
        cap.at[k,'Route'] = i
        cap.at[k,"Exceed_Diminish"] = 0
        cap.at[k,"Count_exceeded_diminished"] = buscap[i]-fullcount[i]
        k+=1

  return cap
```

```python
def aggregate():
  df = pd.read_csv('/content/Dem - Sheet1.csv')
  df["From_loc"] = df["From_loc"]. str. lower()
  df["To_loc"] = df["To_loc"]. str. lower()
  b = pd.DataFrame(df.loc[:,"From_loc"] + " " + df.loc[:,"To_loc"] + " " + df.loc[:,"timing"] + " " + df.loc[:,"date"])
  a = b.iloc[:,0].unique()
  #print(a)
  dicti = {}
  for i in a :
    if i not in dicti :
      dicti[i] = 0
  #print(dicti)

  for i in range(len(b)):
    for j in a:
      if df['From_loc'][i] in j[0:len(df['From_loc'][i])] and df['To_loc'][i] in j and df['timing'][i] in j and df['date'][i] in j:
        dicti[j]+=1
  #print(dicti)

  df.drop_duplicates(inplace=True)
  df.reset_index(drop= True, inplace=True)
  df["number_of_tickets"]= np.zeros(len(df))

  for i in range(len(df)):
    for j in dicti:
      if df['From_loc'][i] in j[0:len(df['From_loc'][i])] and df['To_loc'][i] in j and df['timing'][i] in j and df['date'][i] in j:
        df.at[i,"number_of_tickets"] = dicti[j]
  
  df.to_csv('ticketcount.csv', index=False)

  #Capacity check
  data = pd.read_csv('ticketcount.csv')
  buscount = pd.read_csv('/content/buscount.csv')

  #Count of ticket and route to a dictionary
  fullcount = {}
  for i in range(len(data)):
    f = data['From_loc'][i]+" "+data['To_loc'][i]
    if f not in fullcount:
      fullcount[f] = data['number_of_tickets'][i]
    else:
      fullcount[f]+=data['number_of_tickets'][i]
  #print(fullcount)

  #Dictionary with route name and capacity of passengers
  buscap = {}
  for i in fullcount:
    k = 0
    for j in buscount['Route']:
      k+=1
      if i == j:
        buscap[i] = buscount['Count'][k]*40
  #print(buscap)

  column_names = ["Route", "Exceed_Diminish", "Count_exceeded_diminished"]
  cap = pd.DataFrame(columns = column_names)
  #print(cap)

  #Capacity exceeded or diminished
  for i in buscap:
    if buscap[i] < fullcount[i]:
      print("Exceeded passenger Count by ", fullcount[i]-buscap[i]," in ",i," route")
    else:
      if fullcount[i] < buscap[i]*0.2 :
        print("Passenger Count below by ", buscap[i]-fullcount[i] ," in ",i," route")
  #File containing passenger fall or rise rate
  k=0
  for i in buscap:
    if buscap[i] < fullcount[i]:
      cap.at[k,'Route'] = i
      cap.at[k,"Exceed_Diminish"] = 1
      cap.at[k,"Count_exceeded_diminished"] = fullcount[i]-buscap[i]
      k+=1
    else:
      if fullcount[i] < buscap[i]*0.2 :
        cap.at[k,'Route'] = i
        cap.at[k,"Exceed_Diminish"] = 0
        cap.at[k,"Count_exceeded_diminished"] = buscap[i]-fullcount[i]
        k+=1

  return cap
```

```python
df_train = pd.read_csv('ticketcount.csv')

#list some records from the dataframe
df_train.head()
```

```python
df_train["date"] = pd.to_datetime(df_train["date"],dayfirst=True ,infer_datetime_format=True)
df_train["travel_dow"] = df_train["date"].dt.dayofweek #change the full date to day of week
df_train["travel_month"] = df_train["date"].dt.month #extract month
#df_train['hour_booked'] = pd.to_numeric(df_train['travel_time'].str.extract(r'(^\d*)').loc[:,0])

df_train['ToM'] = df_train.date.dt.day
df_train["isEndWeek"] = df_train["travel_dow"].apply(lambda x: 1 if x >= 5 else 0)
#df_train["isMorning"] = df_train["hour_booked"].apply(lambda x: 1 if x <= 21 else 0)
df_train["isEndMonth"] = df_train["ToM"].apply(lambda x: 1 if (x >= 20) else 0)
```

```python
df_train.drop(['date','ToM'], axis=1, inplace=True)
```

```python
x = df_train.drop(["number_of_tickets"], axis=1)
y = df_train.number_of_tickets
```
